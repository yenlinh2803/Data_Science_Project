```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
user = pd.read_csv('RFReA.csv',sep=',')
user.head(10)
```

```python
import datetime as dt
from datetime import date
NOW = pd.to_datetime(dt.datetime.now())
user['last_activity'] = pd.to_datetime(user['last_activity'])
user['register_date'] = pd.to_datetime(user['register_date'])
usersegment = user 
#abs((end_date-start_date).days)
usersegment['recency'] = NOW - usersegment['last_activity'] # Recency
usersegment['recency'] = usersegment['recency'].astype(datetime)
usersegment.head()
```

```python
 US = user.groupby('user_id').agg({'last_activity': lambda x: (NOW - x.max()).days, # Recency
                                        'fre_cmt': lambda x: x,      # Frequency
                                        'user_age': lambda x: x}) # Monetary Value

US['last_activity'] = US['last_activity'].astype(datetime)

US.head()
```

```python
US['user_id'] = US['last_activity'].astype(int)
```
