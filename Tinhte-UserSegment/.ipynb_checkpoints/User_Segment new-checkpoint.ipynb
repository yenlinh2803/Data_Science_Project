```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
user = pd.read_csv('last_active&cmt&thread.csv',sep=',')

user.head(10)
```

```python
usersegment = pd.DataFrame(data=user, columns=['user_id','username','register_year','user_age','recency','fre_cmt','count_thread'])
usersegment['recency'] = abs(usersegment['recency'])
usersegment.head()
```

```python
quantiles = usersegment.quantile(q=[0.25,0.5,0.75])
quantiles
```

```python
quantiles = quantiles.to_dict()
quantiles
```

```python
USegmentation= usersegment
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def FreThre_Class(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
    
def Re_Class(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
    
def Age_Class(x,p,d):
    if x <= d[p][0.25]:
        return 'A'
    elif x <= d[p][0.50]:
        return 'B'
    elif x <= d[p][0.75]: 
        return 'C'
    else:
        return 'D'
    
```

```python
'''
USegmentation= usersegment
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def FreThre_Class(x,p,d):
    if x <= d[p][0.5]:
        return 0
    else:
        return 1
    
def Re_Class(x,p,d):
    if x <= d[p][0.5]:
        return 1
    else:
        return 0
    
def Age_Class(x,p,d):
    if x <= d[p][0.5]:
        return 'A'
    else:
        return 'B'
'''    
```

```python
USegmentation['ReYe_Quartile'] = USegmentation['register_year'].apply(Age_Class, args=('register_year',quantiles,))
USegmentation['Age_Quartile'] = USegmentation['user_age'].apply(Age_Class, args=('user_age',quantiles,))
USegmentation['Recency_Quartile'] = USegmentation['recency'].apply(Re_Class, args=('recency',quantiles,))
USegmentation['Frequency_Quartile'] = USegmentation['fre_cmt'].apply(FreThre_Class, args=('fre_cmt',quantiles,))
USegmentation['Thread_Quartile'] = USegmentation['count_thread'].apply(FreThre_Class, args=('count_thread',quantiles,))


USegmentation.head(10)
```

```python
USegmentation['Class_1'] = USegmentation.Recency_Quartile.map(str) \
                            + USegmentation.Frequency_Quartile.map(str) \
                            + USegmentation.Thread_Quartile.map(str)

USegmentation['Class_1'] = USegmentation['Class_1'].astype(int)

USegmentation['Class_2'] = USegmentation.Recency_Quartile.map(str) \
                            + USegmentation.Frequency_Quartile.map(str) \
                            + USegmentation.Thread_Quartile.map(str)\
                            + USegmentation.Age_Quartile.map(str)\
                            + USegmentation.ReYe_Quartile.map(str)
USegmentation=USegmentation.sort_values(by=['Class_2'], ascending=True)
USegmentation.head(10)
```

```python
USegmentation[USegmentation['Class_1']==111].sort_values(by=['fre_cmt','recency'], ascending=False).head(10)
```

```python
# Uncomment any of the following lines to: copy data to clipboard or save it to a CSV file.
USegmentation.to_clipboard()
USegmentation.to_csv('TinhTe-USegmentation.csv', sep=',')
```

```python
count_cl_2 = USegmentation.groupby(['Class_1','Class_2']).agg({'user_id': lambda x: x.count()}) 
count_cl_2['user_id'] = count_cl_2['user_id'].astype(int)
count_cl_2.rename(columns={'user_id': 'nb_user'}, inplace=True)
count_cl_2 = count_cl_2.sort_values(by=['nb_user'], ascending = False)
count_cl_2.head(10)
```

```python
count_cl_1 = USegmentation.groupby(['Class_1']).agg({'user_id': lambda x: x.count()}) 
count_cl_1['user_id'] = count_cl_1['user_id'].astype(int)
count_cl_1.rename(columns={'user_id': 'nb_user'}, inplace=True)
count_cl_1 = count_cl_1.sort_values(by=['nb_user'], ascending = False)
count_cl_1
```
